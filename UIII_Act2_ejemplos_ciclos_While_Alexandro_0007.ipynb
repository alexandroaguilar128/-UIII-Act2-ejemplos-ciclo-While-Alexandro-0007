### Ejemplos con ciclo while

### Ejemplo 1

```python
# Contador de números personalizado
# Este programa permite al usuario decidir cuántos números desea imprimir en pantalla, empezando desde el uno.
limite = int(input("¿Hasta qué número quieres contar? ")) # Solicita un número entero al usuario.
i = 1                                                    # Inicializa la variable de control 'i' en 1.
while i <= limite:                                       # Repite mientras 'i' sea menor o igual al límite ingresado.
    print(i)                                             # Muestra el valor actual de 'i'.
    i += 1                                               # Incrementa 'i' en 1 en cada vuelta del ciclo. [4]

print("Programa realizado por Alexandro NL = 1")
```

### Ejemplo 2

```python
# 2. Suma acumulativa con un "centinela"
# En este ejemplo, el ciclo utiliza un valor centinela (un número negativo) para detener la entrada de datos y mostrar la suma total de los números positivos ingresados
suma = 0                                                 # Variable para acumular la suma total.
while True:                                              # Crea un bucle que corre indefinidamente hasta encontrar un 'break'. [5]
    numero = int(input("Ingresa un número (negativo para salir): ")) # Pide un dato al usuario.
    if numero < 0:                                       # Verifica si el número es negativo para terminar el ciclo. [5]
        break                                            # Rompe y sale del bucle 'while' inmediatamente. [5]
    suma += numero                                       # Suma el número ingresado al acumulador. [5]
print(f"La suma total es: {suma}")                       # Imprime el resultado final del cálculo. [5]
print("Programa realizado por Alexandro NL = 1")
```

### Ejemplo 3

```python
# 3. Juego de adivinar el número secreto
# El programa solicita al usuario que adivine un número predefinido y no se detiene hasta que el usuario acierte la respuesta correcta
secreto = 7                                              # Define el número que el usuario debe adivinar. [7]
intento = 0                                              # Variable para almacenar la respuesta del usuario.
while intento != secreto:                                # Repite mientras el intento sea diferente al número secreto. [7]
    intento = int(input("Adivina el número secreto: "))  # Solicita una nueva respuesta en cada iteración.
    if intento != secreto:                               # Si no acierta, envía un mensaje de pista.
        print("Incorrecto, intenta de nuevo.")
print("¡Felicidades! Adivinaste el número.")             # Se ejecuta al salir del ciclo por acertar. [7]
print("Programa realizado por Alexandro NL = 1")
```

### Ejemplo 4

```python
# 4. Validación de entrada de datos (Mes del año)
# Este código es útil para asegurar que el usuario ingrese información válida (en este caso, un mes entre 1 y 12), rechazando cualquier otro valor
mes = 0                                                  # Inicializa la variable con un valor fuera del rango válido.
while mes < 1 or mes > 12:                               # El ciclo sigue mientras el dato sea inválido (menor a 1 o mayor a 12). [9]
    mes = int(input("Introduce el número de un mes (1-12): ")) # Pide al usuario el mes.
    if mes < 1 or mes > 12:                              # Verifica si el dato está fuera del rango. [9]
        print("Valor inválido. Debe estar entre 1 y 12.") # Informa al usuario sobre el error. [9]
print(f"Has seleccionado el mes: {mes}")                 # Confirma el valor correcto una vez aceptado.
print("Programa realizado por Alexandro NL = 1")
```

### Ejemplo 5

```python
# 5. Cálculo del Factorial de un número
# El ciclo while es ideal para cálculos matemáticos decrecientes, como el factorial, donde multiplicamos un número por todos sus antecesores hasta llegar a uno
n = int(input("Ingresa un número para calcular su factorial: ")) # Solicita el número base.
factorial = 1                                            # Inicializa el resultado en 1 (base de la multiplicación). [5]
print(f"Calculando el factorial de {n}...")
while n > 0:                                             # Repite mientras el número sea mayor que cero. [5]
    factorial *= n                                       # Multiplica el acumulado por el valor actual de 'n'. [5]
    n -= 1                                               # Reduce 'n' en una unidad para la siguiente iteración. [5]
print(f"El resultado es: {factorial}")                   # Muestra el resultado final del factorial. [5]
print("Programa realizado por Alexandro NL = 1")
```
